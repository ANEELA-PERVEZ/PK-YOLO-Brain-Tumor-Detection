```python
import pandas as pd
import os

# 1. Folders ke paths define karein
result_folders = {
    "Axial": "/content/drive/MyDrive/PK-YOLO-main/runs/train/exp13",
    "Coronal": "/content/drive/MyDrive/PK-YOLO-main/runs/train/coronal_scratch_final",
    "Sagittal": "/content/drive/MyDrive/PK-YOLO-main/runs/train/sagittal_scratch_final"
}

# Paper benchmarks for comparison
paper_benchmarks = {
    "Axial": 0.947,
    "Coronal": 0.805,
    "Sagittal": 0.582
}

final_results = []

print("🔄 Results extract ho rahe hain...\n")

for plane, folder in result_folders.items():
    csv_path = os.path.join(folder, 'results.csv')

    if os.path.exists(csv_path):
        df = pd.read_csv(csv_path)
        df.columns = df.columns.str.strip() # Hidden spaces khatam karein

        # Aapke specific columns map karein
        m50_col = 'metrics/mAP_0.5'
        prec_col = 'metrics/precision'
        rec_col = 'metrics/recall'
        m95_col = 'metrics/mAP_0.5:0.95'

        if m50_col in df.columns:
            # Best epoch based on mAP_0.5
            best_row = df.loc[df[m50_col].idxmax()]

            # 1. Obtained data (Humara Kaam)
            final_results.append({
                "MRI Plane": plane,
                "Source": "Obtained (Ours)",
                "Precision": round(best_row[prec_col], 3),
                "Recall": round(best_row[rec_col], 3),
                "mAP50": round(best_row[m50_col], 3),
                "mAP50:95": round(best_row[m95_col], 3)
            })

            # 2. Paper data (Comparison ke liye)
            final_results.append({
                "MRI Plane": plane,
                "Source": "Paper (PK-YOLO)",
                "Precision": "-",
                "Recall": "-",
                "mAP50": paper_benchmarks[plane],
                "mAP50:95": "-"
            })
        else:
            print(f"❌ {plane}: Column '{m50_col}' nahi mila.")
    else:
        print(f"⚠️ Warning: {plane} ka folder nahi mila.")

# Results Table Format mein Print karein
if final_results:
    final_df = pd.DataFrame(final_results)
    # Sorting for professional view
    final_df['PlaneOrder'] = final_df['MRI Plane'].map({"Axial": 1, "Coronal": 2, "Sagittal": 3})
    final_df = final_df.sort_values(['PlaneOrder', 'Source'], ascending=[True, False]).drop('PlaneOrder', axis=1)

    print("\n✅ FINAL COMPARATIVE ANALYSIS TABLE:\n")
    print(final_df.to_string(index=False))
else:
    print("❌ Data extract nahi ho saka.")
```

```python
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pytorch_grad_cam.utils.image import show_cam_on_image
import sys
sys.path.insert(0, '/content/drive/MyDrive/PK-YOLO-main')
from models.common import DetectMultiBackend

# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Model
model = DetectMultiBackend(
    '/content/drive/MyDrive/PK-YOLO-main/runs/kfold_axial_t1wce_2_class/fold_5_run/weights/best.pt',
    device=device,
    fuse=False
)
model.eval()
for param in model.parameters():
    param.requires_grad = True

# Image load
img_path = '/content/drive/MyDrive/PK-YOLO-main/datasets/brain_tumor/sagittal_t1wce_2_class/images/test/00000_102.jpg'
img = cv2.imread(img_path)
img = cv2.resize(img, (640, 640))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_float = np.float32(img_rgb) / 255.0
input_tensor = torch.from_numpy(img_float).permute(2,0,1).unsqueeze(0).to(device)
input_tensor.requires_grad = True

# Manual hooks — GradCAM library bypass ✅
activations = []
gradients = []

def forward_hook(module, input, output):
    activations.append(output)

def backward_hook(module, grad_input, grad_output):
    gradients.append(grad_output[0])

target_layer = model.model.model[30]
h1 = target_layer.register_forward_hook(forward_hook)
h2 = target_layer.register_full_backward_hook(backward_hook)

# Forward pass
with torch.enable_grad():
    output = model(input_tensor)

# Score extract
if isinstance(output, (list, tuple)):
    out = output[0]
    if isinstance(out, (list, tuple)):
        out = out[0]
score = out[..., 4].max()

# Backward pass
model.zero_grad()
score.backward()

# Remove hooks
h1.remove()
h2.remove()

# GradCAM compute
act = activations[0].detach().cpu().numpy()[0]  # [C, H, W]
grad = gradients[0].detach().cpu().numpy()[0]   # [C, H, W]
weights = np.mean(grad, axis=(1, 2))             # [C]
cam_map = np.zeros(act.shape[1:], dtype=np.float32)
for i, w in enumerate(weights):
    cam_map += w * act[i]

cam_map = np.maximum(cam_map, 0)
cam_map = cv2.resize(cam_map, (640, 640))
cam_map = (cam_map - cam_map.min()) / (cam_map.max() - cam_map.min() + 1e-8)

# Visualize
visualization = show_cam_on_image(img_float, cam_map, use_rgb=True)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1); plt.imshow(img_rgb); plt.title('Original MRI')
plt.subplot(1,2,2); plt.imshow(visualization); plt.title('GradCAM Heatmap')
plt.savefig('/content/drive/MyDrive/gradcam_output.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ GradCAM saved!")
```
